# **EXPLORATORY DATA ANALYSIS:-**

EDA stands for Exploratory Data Analysis. It is a crucial step in the data analysis process that involves examining, cleaning, summarizing, and visualizing data to gain insights, discover patterns, detect anomalies, and formulate hypotheses. EDA helps analysts and data scientists understand the underlying structure of the data, identify relationships between variables, and prepare the data for further modeling or analysis.

Here's a detailed explanation of the key steps involved in Exploratory Data Analysis:

**Data Collection:**
The first step is to gather data from various sources, such as databases, files, APIs, or web scraping. The data can be structured (e.g., in a tabular format like CSV or Excel) or unstructured (e.g., text, images, videos).

**Data Cleaning:**
Data cleaning is the process of identifying and correcting errors, inconsistencies, and inaccuracies in the dataset. This step involves handling missing values, removing duplicates, and dealing with outliers.

**Data Exploration:**
In this step, analysts explore the data to get an initial understanding of its structure and characteristics. Descriptive statistics, such as mean, median, mode, standard deviation, and quartiles, are computed for numerical variables. For categorical variables, frequency counts and unique values are examined.

**Data Visualization:**
Data visualization is a powerful technique to present the data graphically. Various types of plots and charts, such as bar plots, line plots, scatter plots, histograms, box plots, heatmaps, and pie charts, are used to visualize the distribution, relationships, and trends in the data.

**Feature Engineering:**
Feature engineering involves creating new features from the existing ones to enhance the predictive power of the data. It may include transforming variables, scaling data, or encoding categorical variables.

**Hypothesis Testing:**
Hypothesis testing is performed to validate or reject assumptions about the data. Statistical tests are used to determine if the observed patterns or differences are significant or occurred by chance.

**Data Transformation:**
Data transformation includes standardization, normalization, or log transformations to ensure that data follows certain statistical assumptions required for modeling.

**Correlation and Association Analysis:**
Correlation analysis helps identify relationships between numerical variables, while association analysis finds patterns and connections in categorical data using techniques like correlation matrices and association rules (e.g., Apriori algorithm).

**Outlier Detection:**
Outliers are extreme data points that deviate significantly from the rest of the data. Identifying and handling outliers is crucial as they can adversely affect the model performance.

**Summarizing Findings:**
After performing various analyses and visualizations, analysts summarize their findings and insights about the data. They may make recommendations or outline the next steps for further analysis or modeling.

# **BASIC IDEA OF EDA**

```python
# libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Example data
data = {
    "Brands": ["Samsung S8", "Samsung S20", "Realme 5i", "Iphone 14", "Realme C33", "Iphone 11"],
    "Year_launched": ["2010", "2021", "2020", "2022", "2023", "2017"],
    "Users": ["200000", "750000", "1500000", "2000000", "150000", "300000"],
    "Prices": ["15000", "200000", "25000", "350000", "30000", "80000"],
}
```

```python
# Create a DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Convert numerical columns to numeric data types
df["Year_launched"] = df["Year_launched"].astype(int)
df["Users"] = df["Users"].astype(int)
df["Prices"] = df["Prices"].astype(int)
```

```python
# 1. Aggregation: Average number of users and prices per brand
aggregated_data = df.groupby("Brands").agg({"Users": "mean", "Prices": "mean"}).reset_index()
print("\nAverage Number of Users and Prices per Brand:")
print(aggregated_data)
```

# **DATA VISUALIZATION**

```python
# Visualization 1: Bar plot - Number of mobile phones per brand
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="Brands", palette="viridis")
plt.xlabel("Brands")
plt.ylabel("Count")
plt.title("Number of Mobile Phones per Brand")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 2: Line plot - Number of users vs. Year launched
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Year_launched", y="Users", marker='o', color='b')
plt.xlabel("Year Launched")
plt.ylabel("Number of Users")
plt.title("Number of Users vs. Year Launched")
plt.grid(True)
plt.show()
```

```python
# Visualization 3: Scatter plot - Price vs. Number of Users
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Users", y="Prices", hue="Brands", s=100, palette="Set2")
plt.xlabel("Number of Users")
plt.ylabel("Price")
plt.title("Price vs. Number of Users")
plt.grid(True)
plt.show()
```

```python
# Visualization 4: Box plot - Prices distribution
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Brands", y="Prices", palette="pastel")
plt.xlabel("Brands")
plt.ylabel("Price")
plt.title("Prices Distribution by Brand")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 5: Bar plot - Average price per brand
average_prices = df.groupby("Brands")["Prices"].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=average_prices, x="Brands", y="Prices", palette="magma")
plt.xlabel("Brands")
plt.ylabel("Average Price")
plt.title("Average Price per Brand")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 6: Pie chart - Market share of each brand
plt.figure(figsize=(8, 8))
plt.pie(df["Users"], labels=df["Brands"], autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3"))
plt.axis('equal')
plt.title("Market Share of Each Brand")
plt.show()
```

```python
# Visualization 7: Histogram - Distribution of the number of users
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="Users", bins=10, kde=True, color='purple')
plt.xlabel("Number of Users")
plt.ylabel("Frequency")
plt.title("Distribution of the Number of Users")
plt.show()
```

```python
# Visualization 8: Pair plot - Relationships between numerical variables
sns.pairplot(df, hue="Brands", palette="Set2")
plt.show()
```

```python
# Visualization 9: Heatmap - Correlation between numerical variables
plt.figure(figsize=(8, 6))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Visualization 10: Violin plot - Price distribution per brand
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x="Brands", y="Prices", palette="Set1")
plt.xlabel("Brands")
plt.ylabel("Price")
plt.title("Price Distribution per Brand")
plt.xticks(rotation=45)
plt.show()
```

# **WORD2VEC**

Word2Vec is a popular word embedding technique used in natural language processing (NLP) and machine learning to convert words or phrases into numerical vectors (word embeddings). It was introduced by Tomas Mikolov et al. at Google in 2013. Word embeddings help represent words in a continuous vector space, allowing algorithms to understand and work with semantic relationships between words.

The basic idea behind Word2Vec is to learn distributed representations of words by training a shallow neural network on a large corpus of text. This network aims to predict a target word based on its neighboring words (context) or vice versa. Word2Vec creates a dense representation for each word, capturing semantic similarities, analogies, and relationships between words.

# There are two main architectures in Word2Vec:

**Continuous Bag of Words (CBOW):**
CBOW predicts a target word from its surrounding context words.
The input to the model is the context words (source), and the output is the target word (target).
CBOW is computationally efficient and performs well on smaller datasets.

**Skip-Gram:**
Skip-Gram predicts surrounding context words given a target word.
The input to the model is the target word, and the output is the context words.
Skip-Gram is more effective for capturing rare words and works well with large datasets.
The training process of Word2Vec involves updating the word embeddings based on the model's performance in predicting the context words or target words. The objective is to learn word representations that capture semantic relationships and co-occurrence patterns in the training data.

**The resulting word embeddings from Word2Vec have some interesting properties:**

Similar words tend to have similar vector representations in the embedding space.
Semantic relationships between words are captured through vector arithmetic. For example, embedding("king") - embedding("man") + embedding("woman") may result in a vector close to embedding("queen").
Word embeddings can be used to initialize word-related tasks, such as sentiment analysis, named entity recognition, and machine translation.
Word2Vec has revolutionized the field of NLP and enabled significant advancements in various natural language understanding tasks. The word embeddings learned through Word2Vec have been widely used as a foundational step in building more complex NLP models and extracting meaningful features from text data.

```python
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
```

```python
# Add random data for Word2vec Visualization
brand_text_data = {
    "Samsung S8": "Great camera quality and long battery life.",
    "Samsung S20": "Super AMOLED display and powerful processor.",
    "Realme 5i": "Budget-friendly phone with a large screen.",
    "Iphone 14": "The latest iPhone with advanced features.",
    "Realme C33": "Decent camera and good performance.",
    "Iphone 11": "A popular iPhone model with excellent battery life."
}
```

```python
# Create Word2Vec embeddings
sentences = [text.split() for text in brand_text_data.values()]
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
```

```python
# Reduce dimensions for visualization using PCA
X = model.wv[model.wv.key_to_index]
pca = PCA(n_components=2)
result = pca.fit_transform(X)
```

```python
# Create a DataFrame for Word2Vec visualization
word2vec_df = pd.DataFrame(result, index=model.wv.key_to_index, columns=["x", "y"])
```

```python
# Visualization 11: Word2Vec Visualization
plt.figure(figsize=(10, 8))
sns.scatterplot(data=word2vec_df, x="x", y="y", hue=word2vec_df.index, palette="Set1", s=150)
for word, (x, y) in word2vec_df.iterrows():
    plt.text(x, y, word, fontsize=10, ha='center', va='center')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Word2Vec Visualization for Mobile Phone Brands")
plt.grid(True)
plt.legend(title="Brands", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
