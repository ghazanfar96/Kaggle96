```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
sns.set_style("dark")
```

```python
path = "/kaggle/input/fivethirtyeight-comic-characters-dataset/marvel-wikia-data.csv"
comic = pd.read_csv(path)
```

# **LINEPLOT**

```python
plt.figure(figsize=(16,6))

sns.lineplot(data=comic)
```

# **BARPLOT**

```python
plt.figure(figsize=(12,6))
sns.barplot(x=comic['SEX'], y=comic['Year'])
```

# **HISTOGRAM**

```python
plt.figure(figsize=(12,6))
sns.histplot(comic['Year'])
```

# **SCATTERPLOT**

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=comic['APPEARANCES'], y=comic['Year'], hue=comic['SEX'])
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=comic['SEX'], y=comic['Year'], hue=comic['HAIR'])
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x=comic['ALIVE'], y=comic['Year'], hue=comic['EYE'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=comic['ALIGN'], y=comic['Year'], hue=comic['SEX'])
```

# **SCATTERPLOT WITH REGRESSION LINE**

```python
plt.figure(figsize=(12,6))
sns.regplot(x=comic['APPEARANCES'], y=comic['Year'])
```

# **COLOR-CODED SCATTERPLOT**

```python
plt.figure(figsize=(12,6))
sns.lmplot(x="APPEARANCES", y="Year", hue="ALIGN", data=comic)
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot(x=comic['SEX'], y=comic['Year'])
```

```python
label_encoder = LabelEncoder()
comic['EYE'] = label_encoder.fit_transform(comic['EYE'])
```

```python
pivot_table = comic.pivot_table(index='Year', columns='EYE', aggfunc=len, fill_value=0)
```

# **COMIC HEATMAP**

```python
plt.figure(figsize=(16,8))
sns.heatmap(pivot_table, cmap="coolwarm", annot=True, fmt="d", cbar=False)
```

# **New Data Load**

```python
path = "/kaggle/input/data-for-datavis/cancer.csv"
cancer = pd.read_csv(path)
```

```python
cancer
```

```python
cancer.info()
```

```python
cancer.columns
```

```python
cancer.drop(['Diagnosis'], axis=1)
```

```python
label_encoder = LabelEncoder()
cancer['Diagnosis'] = label_encoder.fit_transform(cancer['Diagnosis'])
```

```python
pivot_table = cancer.pivot_table(index='Id', columns='Diagnosis', aggfunc=len, fill_value=0)
```

# **HEATMAP**

```python
plt.figure(figsize=(16,8))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt="d", cbar=False)
```
