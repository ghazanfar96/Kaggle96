```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Comment this if the data visualisations doesn't work on your side
%matplotlib inline

plt.style.use('bmh')
```

```python
# Ghazanfar Ali
df = pd.read_csv('/kaggle/input/covid-19-nlp-text-classification/Corona_NLP_test.csv')
df.head()
```

```python
# Ghazanfar Ali
df.info()
```

```python
summary_stats = df.describe()
print("Summary Statistics:\n", summary_stats)
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df, x='ScreenName', bins=100, kde=True)
plt.title("Histogram")
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(df, x='ScreenName', y='Sentiment')
plt.title("Boxplot")
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x=df.index, y='Sentiment')
plt.title("Scatter Plot")
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x=df.index, y='ScreenName')
plt.title("Scatter Plot")
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
